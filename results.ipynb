## 3.2 How many containers are exported and imported per shipsystem (SSY)?                    

SSYs are the different ocean services which are connected to voyages. In the data set, the more
generalized term MOT_Service is used (MOT = Mode of Transport)

</a>

```python
quiz_iot_events['DP_VOYAGE_NO'].nunique()
```

```python
quiz_voyages['DP_VOYAGE_NO'].nunique()
```

```python
col_filt=['UnitNumber_masked','DP_VOYAGE_NO']
quiz_iot_k=quiz_iot_events[quiz_iot_events.columns[quiz_iot_events.columns.isin(col_filt)]]
quiz_iot_k.drop_duplicates (subset=['UnitNumber_masked','DP_VOYAGE_NO'], inplace=True)  
quiz_iot_k.reset_index(drop=True, inplace=True)
quiz_iot_k.head(10)
```

```python
quiz_iot_voyage=quiz_iot_k.merge(quiz_voyages, on='DP_VOYAGE_NO',how='left')
```

```python
quiz_iot_voyage_g=quiz_iot_voyage.groupby("MOT_SERVICE")['UnitNumber_masked'].count()
quiz_iot_voyage.shape
```

```python
print(f"Containers distribution per shipsystem (SSY): {(quiz_iot_voyage_g)}")
```

## 3.4 What is the maximum distance a container traveled from the hinterland to a port (and vice versa)??

</a>

```python
quiz_iot_events['Distance'] = np.where(quiz_iot_events['UnitNumber_masked']==quiz_iot_events['UnitNumber_masked'].shift(),
                         haversine(quiz_iot_events['General_Pos_Lat'].shift(), 
                                   quiz_iot_events['General_Pos_Long'].shift(), 
                                   quiz_iot_events['General_Pos_Lat'], quiz_iot_events['General_Pos_Long']
                                  ),0)
```

```python
quiz_iot_events['Distance'].max()
print(f"Maximum distance without stop: {quiz_iot_events['Distance'].max()}")
```

```python

quiz_iot_events[quiz_iot_events['Distance']==2811.976161317974]
val_df=quiz_iot_events[quiz_iot_events['UnitNumber_masked']=='70e71af3efb0137fe7769d73ba460072']
```

```python
q3_val = folium.Map()
routes_map(val_df,q3_val)
```

```python
df_sum=quiz_iot_events.groupby("UnitNumber_masked")['Distance'].sum().reset_index()
df_sum['Distance'].max()
```

```python
df_sum[df_sum['Distance']==8833.813512451317]
val_df2=quiz_iot_events[(quiz_iot_events['UnitNumber_masked']=='2fe646a92d58075097c8c27dd9da50fe')
&(~quiz_iot_events['type_of_movement'].isin(['stay','moving']))
].reset_index()

val_df2['Distance'] = np.where(val_df2['UnitNumber_masked']==val_df2['UnitNumber_masked'].shift(),
                         haversine(val_df2['General_Pos_Lat'].shift(), 
                                   val_df2['General_Pos_Long'].shift(), 
                                   val_df2['General_Pos_Lat'], val_df2['General_Pos_Long']
                                  ),0)
val_df2=val_df2[val_df2['type_of_movement']=='entering pos']
```
